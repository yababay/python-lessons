# Переменные в Python

Объявение переменных - одна из самых понятных операций в Python. Она осуществляется с помощью знака `=`, слева от которого указывают имя переменной¸справа - ее значение, т.е. так же, как в математике:

```python
name='Гурбангулы Мяликгулыевич Бердымухамедов'
age=62  # знак "=" может быть окружен пробелами

print(name) # функция print - для вывода значений на экран
print(age)
```

В отличие от многих других языков программирования, Python не требует ключевых слов (var, let, const) для объявления переменных, а так же не заботится о типе (число, строка, объект), который будет в ней храниться. Тип определяет сам язык на основе указанного справа от знака равенства выражения. Если оно состоит из цифр - значит это число, если заключено в кавычки - значит строка, если заключено в квадратные скобки - значит список и т.п. 

Переопределение типа переменной допускается, но не приветствуется:

```python
age = 'sixty two'
print(age)
```

В Python запрещены имена переменных, начинающиеся с цифр (`62age`), служебных символов (`!name`, `?city`), они не должны содержать пробелов (`last name`). Допускается использование в именах переменных символа `_` ("нижнее подчеркивание").

Имена переменных в Python регистрозависимы, т.е. `name`, `NAME`, `Name` - это разные переменные.

При именовании переменных следует соблюдать и некоторые негласные правила. Эти соглашения не закреплены в стандартах, но они необходимы для взаимопонимания между разработчиками. Поскольку современное программирование - это программирование коллективное, соблюдение отраслевого этикета играет в профессии важную роль. 

1. избегать ничего не значащих переменных, таких, как `a`, `b`, `x`, `y`;
2. использовать в именах переменных маленькие буквы латинского алфавита: `name`, `age`; применение национальных букв допускается, если вы уверены, что ваша программа будет преимущественно развиваться и использоваться среди программистов-соотечественников: `имя`, `возраст`;
3. при необходимости введения переменных, состоящих из нескольких слов, следует использовать "змеиный стиль", т.е. соединять слова с помощью знака `_` (`last_name`, `place_of_birth`).
4. заглавные буквы используются в особых случаях, например, для именования констант, классов; об этом речь пойдет позже.
